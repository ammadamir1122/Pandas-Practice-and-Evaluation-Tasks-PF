## 1. Which month saw the highest joining of the employees

   function name: max_joinings_month()
   
   inputs: DataFrame
   
   outputs: str

```python
def max_joinings_month(df):
    max_joining_month =df.groupby('Month Name of Joining')['Emp ID'].count()
    max=  max_joining_month.idxmax()
    return max

print("on demo.csv",max_joinings_month(df))
# print("on test.csv",max_joinings_month(test_df))
```

## 2. What is the mean salary is the state KS
   
   function name: mean_ks_salary()
   
   inputs: DataFrame
   
   outputs: float

```python
def mean_ks_salary(df):
    mean_salary_ks = df.groupby('State')['Salary'].mean()
    return mean_salary_ks.loc['KS']

print("on demo.csv",mean_ks_salary(df))
# print("on test.csv",mean_ks_salary(test_df))
```

## 3. Which Region has highest mean salary
   
   function name: max_mean_salary_region()
   
   inputs: DataFrame
   
   outputs: str

```python
def max_mean_salary_region(df):
    region_salary_means = df.groupby("Region")["Salary"].mean()
    region_with_highest_salary = region_salary_means.idxmax()
    return region_with_highest_salary

print("on demo.csv",max_mean_salary_region(df))
# print("on test.csv",max_mean_salary_region(test_df))
```

## 4. Which region has lowest mean salary
   
   function name: min_mean_salary_region()
   
   inputs: DataFrame
   
   outputs: str

```python
def min_mean_salary_region(df):
    region_salary_means = df.groupby("Region")["Salary"].mean()
    region_with_lowest_salary = region_salary_means.idxmin()
    return region_with_lowest_salary

print("on demo.csv",min_mean_salary_region(df))
# print("on test.csv",min_mean_salary_region(test_df))
```

## 5. Which City has highest mean weight
   
   function name: max_mean_weight_city()
   
   inputs: DataFrame
   
   outputs: str

```python
def max_mean_weight_city(df):
    city_weight_means = df.groupby("City")["Weight in Kgs."].mean()
    city_with_highest_weight = city_weight_means.idxmax()
    return city_with_highest_weight

print("on demo.csv",max_mean_weight_city(df))
# print("on test.csv",max_mean_weight_city(test_df))
```

## 6. What is the difference between mean salary of Doctors and Professors
   
   function name: dr_prof_mean_salary_diff()
   
   inputs: DataFrame
   
   outputs: float

```python
def dr_prof_mean_salary_diff(df):
    doctors_mask = df["Name Prefix"].isin(["Dr.", "Drs."])
    professors_mask = df["Name Prefix"] == "Prof."
    doctors_salaries = df.loc[doctors_mask, "Salary"]
    professors_salaries = df.loc[professors_mask, "Salary"]
    
    # Calculate the mean salary for each group
    doctors_mean_salary = doctors_salaries.mean()
    professors_mean_salary = professors_salaries.mean()
    
    # Calculate the difference between the mean salary of doctors and professors
    salary_difference = professors_mean_salary-doctors_mean_salary 
    
    return salary_difference

print("on demo.csv",dr_prof_mean_salary_diff(df))
# print("on test.csv",dr_prof_mean_salary_diff(test_df))
```

## 7. How many female use yahoo
   
   function name: females_with_yahoo()
   
   inputs: DataFrame
   
   outputs: int

```python
def females_with_yahoo(df):
    females_mask = df["Gender"] == "F"
    yahoo_mask = df["E Mail"].str.contains("yahoo")
    females_with_yahoo_emails = df.loc[females_mask & yahoo_mask]
    
    # Count the number of females with Yahoo email addresses
    count = len(females_with_yahoo_emails)
    
    return count

print("on demo.csv",females_with_yahoo(df))
# print("on test.csv",females_with_yahoo(test_df))
```
